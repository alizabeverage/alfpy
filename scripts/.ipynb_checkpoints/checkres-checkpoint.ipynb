```python
import pickle
from dynesty import plotting as dyplot

res1 = pickle.load(open('../test_dynesty_res.p', "rb" ))
use_keys = ['velz', 'sigma', 'logage', 'zh', 'feh', 
            'ah', 'ch', 'nh', 'mgh', 'sih', 'cah', 'tih'] + \
           ['imf1', 'imf2', 'imf3']
ndim = res1.samples.shape[1]

fig, axes = dyplot.traceplot(res1, show_titles=False, trace_cmap='plasma',
                             title_kwargs={'fontsize': 28, 'y': 1.05}, quantiles=[0.005,0.995],
                             fig=plt.subplots(ndim, 2, figsize=(15, ndim*3)))
fig.tight_layout()
for i, ikey in enumerate(use_keys):
    axes[i, 0].set_ylabel(ikey, fontsize=15)
    axes[i, 1].set_xlabel(ikey, fontsize=15) 
    
figdir = '/Users/menggu/work/massive/figures/'
fig.savefig('{0}/alfpy_test1.png'.format(figdir),
             bbox_inches='tight',dpi=85 )
```

```python
lnz_truth = ndim * -np.log(2 * 10.)  # analytic evidence solution
fig, axes = dyplot.runplot(res1, lnz_truth=lnz_truth)  # summary (run) plot
axes[3].set_ylim(-1e-19, 1e-19,)
```
