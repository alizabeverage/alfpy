```python
def getimf(m, imf1=1.3, imf2=2.3, imf3=0.08):
    m1 = imf3
    m2 = 0.5
    m3 = 1.0
    a1 = imf1
    a2 = imf2
    a3 = 2.30   
    
    const3 = -m3**-a3
    const2 = -m3**-a3 - m2**-a2
    if m>=m3:
        return m**-a3
    elif m>=m2:
        return m**-a2 + (m3**-a3 - m3**-a2)
    elif m>=m1:
        return m**-a1 + (m3**-a3 - m3**-a2) + (m2**-a2 - m2**-a1)
    else:
        return np.nan
    
getimf = np.vectorize(getimf)
```
